## Univariate Exploration

> ### What is the employment status of the borrowers?

```python
order = loan.EmploymentStatus.value_counts().index
plt.figure(figsize = (12, 6))
plt.xticks(rotation = 45)
plt.xlabel('Employment Status', size = 12)
plt.ylabel('Count')
plt.title('Employment Status Distribution', size = 14, fontweight='bold')

ax = sns.countplot(x='EmploymentStatus', data = loan,color = 'black', order = order)
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light')


plt.show()
```

The distribution of the employment status provides insight into the composition of borrowers of Prosper Loan. As seen in the employment status distribution, the majority of borrowers are gainfully employed with a sizeable percentage working full-time. Furthermore, the distribution showed that a very small amount (0.70%) of borrowes are retired. With this insight, it can be deduced that the Prosper Loan company are in good business with minimal risk as majority of their borrowers can pay back the loan. 

> ###     What is the distribution of loan statuses within Prosper Loan's portfolio?

```python
order = loan.LoanStatus.value_counts().index
plt.figure(figsize = (12, 6))
plt.xticks(rotation = 90)
plt.xlabel('Loan Status', size = 12)
plt.ylabel('Count')
plt.title('Loan Status Distribution', size = 14, fontweight='bold')

ax = sns.countplot(x='LoanStatus', data = loan,color = 'black', order = order)
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light')


plt.show()
```

This distribution of loan statuses sheds light on how loans are now faring in terms of their ability to be repaid. The distribution shows the number of loans that have defaulted or are past due at different times, it also shows how many have been cancelled, charged off, completed, or are in various stages of repayment. This insight gives understanding to the loan portfolio, locating any potential red flags, such as defaulted loans or loans with high delinquencies.

> ### What is the distribution of Prosper score in the Prosper Loan company?

```python
plt.figure(figsize = (12, 6))
plt.xlabel('Prosper Score', size = 12)
plt.ylabel('Count')
plt.title('Prosper Score Distribution', size = 14, fontweight='bold')

ax = sns.countplot(x='ProsperScore', data = loan,color = 'black')
ax.bar_label(ax.containers[0], fmt='%d',fontsize=9,fontweight='light')


plt.show()
```

The Prosper Score distribution is a custom risk score built using historical Prosper data. The score ranges from 1-11, with 11 being the best, or lowest risk score and 1 being the worst, or highest risk score. The prosper scores shows that the scores are normally distributed.

## Bivariate Exploration

> ### This section investigates relationships between pairs of variables in the Prosper Loan dataset. 

```python
loan.info()
```

> ### Investigating the relationship between Loan status and Employment status

```python
plt.figure(figsize = (10, 6))
sns.countplot(data = loan, x = 'EmploymentStatus', hue = 'LoanStatus', palette = 'Set2')
plt.title('Loan Status by Employment Status', size = 14, fontweight = 'bold')
plt.xlabel('Employment Status', size = 12)
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.legend(title = 'Loan Status',bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

> ### Relationship between Borrower rate and Lender yield

```python
sns.regplot(data = loan,x = 'BorrowerRate', y = 'LenderYield', color = 'black')
plt.title('Relationship Between Borrower Rate and Lender Yield', size = 14, fontweight='light')
plt.show()
```

> ### Relationship between Prosper Rating and Homeownership

```python
plt.figure(figsize = (12, 6))
palette = sns.color_palette(['#000000', '#AAAAAA'])
sns.countplot(data = loan, x = 'ProsperRating (Alpha)', hue = 'IsBorrowerHomeowner', palette = palette)

plt.xlabel("Prosper Rating", size = 12)
plt.ylabel("Count", size = 12)
plt.title('Relationship Between Prosper Rating and Home ownership', size = 14, fontweight='bold')

ax = sns.barplot(x = 'ProsperRating (Alpha)', y = 'IsBorrowerHomeowner', data = loan)
for container in ax.containers:
    ax.bar_label(container, fmt='%d',fontsize=9,fontweight='light')

plt.show()
```
